# Step 4 — Linear Probing Validation

Train a linear probe on emotion activations and evaluate generalization to held-out text.

```python
import sys; from pathlib import Path; sys.path.insert(0, str(Path('..') / 'src'))
from probing import build_probe_dataset, train_linear_probe
from extract_activations import load_stimuli, extract_activation_vectors
import json, numpy as np

# Load per-narrative activations (need per-sample, not mean)
# Re-extract with return_per_sample=True or load from cache
# For now use the mean vectors as a sanity check
from emotion_vectors import load_vectors
matrix, labels = load_vectors()
```

```python
# Synthetic multi-sample dataset: perturb mean vectors with small noise to simulate per-narrative activations
np.random.seed(42)
n_per_emotion = 5
X_rows, y_rows = [], []
for i, label in enumerate(labels):
    for _ in range(n_per_emotion):
        X_rows.append(matrix[i] + np.random.randn(*matrix[i].shape) * 0.01)
        y_rows.append(i)
X = np.stack(X_rows)
y = np.array(y_rows)
print(f'Dataset: X={X.shape}, y={y.shape}')
```

```python
results = train_linear_probe(X, y, cv=5)
print(f"CV accuracy: {results['cv_accuracy_mean']:.3f} ± {results['cv_accuracy_std']:.3f}")
```
