# Step 3 — Emotion Vector Geometry Analysis

Visualize the structure of emotion vectors: PCA map, cosine similarity heatmap, valence/arousal clustering.

```python
import sys; from pathlib import Path; sys.path.insert(0, str(Path('..') / 'src'))
from emotion_vectors import load_vectors, cosine_similarity_matrix, plot_emotion_map, get_nearest_emotions
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

matrix, labels = load_vectors()
print(f'Shape: {matrix.shape}')
```

```python
plot_emotion_map(matrix, labels)
```

```python
# Cosine similarity heatmap (subset for readability)
sim = cosine_similarity_matrix(matrix)
subset = slice(0, 30)
plt.figure(figsize=(12, 10))
sns.heatmap(sim[subset, subset], xticklabels=labels[subset], yticklabels=labels[subset], cmap='coolwarm', center=0)
plt.title('Emotion Vector Cosine Similarities')
plt.tight_layout()
plt.show()
```

```python
# Nearest neighbors for a few anchor emotions
for anchor in ['fear', 'joy', 'anger', 'sadness']:
    if anchor in labels:
        print(f"\nNearest to '{anchor}':")
        for name, score in get_nearest_emotions(anchor, labels, matrix):
            print(f'  {name}: {score:.3f}')
```
